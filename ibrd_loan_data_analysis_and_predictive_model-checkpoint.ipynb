<h1 align="center"> World Bank IBRD Loan Data Analysis and Predictive Model </h1> <br>

The International Bank for Reconstruction and Development [(IBRD)](https://www.worldbank.org/en/who-we-are/ibrd) is the world's largest development cooperative working with the World Bank to service middle or low-income countires. One of the main services is loans for the countries to carry out diverse projects such as construction, energy, road, or education. Historically there have been about 758K cases since the group was formed in 1947 for redevelopment after WWII. Most of these were disbursed and repaid, however, some of them were cancelled. The loan cancellation is a loan that was ready to begin but widthdrawn before disbursement. Considering the scale of loan amounts, cancellations would cause a tremendous waste of time as well as financial cost. For that reason, this project focused on analyzing cancelled loans and building a predictive model to prevent or prepare for such loans.

```python
def recession_years(countries):
    keep_idx = []
    for country in countries:
        country_df = df[df['country']==country][['closed_date']]
        # cancelled loans
        loans = num_cancelled_by_year(country) 
        loans = [loans[0] + x - y for x, y in zip(loans[1:], loans)]
        # recession years
        years = [(loan, year) for loan, year in zip(loans, np.arange(1980, 2024))]
        years = [year[1] for year in years if year[0] != 0]
        print(country, "experienced recessions in", years)
        # indice of loans
        recession = country_df['closed_date'].apply(lambda x : x.year).isin(years)
        idx = country_df[recession].index.tolist()
        keep_idx.append(idx)
    
    keep_idx = [x for y in keep_idx for x in y]
    return keep_idx

def final_dataset(file_name):
    global df
    if not os.path.isfile(file_name):
        # filter recession years by country if not processed
        countries = df['country'].unique()
        ibrd_df_processed = df.loc[recession_years(countries)]
        # write a file
        outfile = open('ibrd_df_processed', 'wb')
        pickle.dump(ibrd_df_processed, outfile)
        outfile.close()
    # load the file in data frame if already processed
    infile = open(file_name, 'rb')
    ibrd_dict = pickle.load(infile)
    df = pd.DataFrame.from_records(ibrd_dict)
    infile.close()
    
    return df
```

```python
def sector_finder(project_name):
    project_name= project_name.split(" ")
    # handel the conventon - country code XX-..
    project_name[0]= re.sub(r"^\w+-", "",project_name[0] )
    # each word has at least 2 lenght
    project_name = list(filter(lambda x: len(x)>1, project_name))
    
    # look up proj_dict to classify the sector
    
    sector = ""
    score = 0
    for word in project_name:
        for key, val in proj_dict.items():
            scores=[fuzz.ratio(word.lower(),proj) for proj in val]
            if max(scores)>score:
                sector = key
                score =max(scores)
        # early stop if matched enough
        if score> 95:
            return sector
    # no sector if not enough
    if score<65:
        return "other"
    
    return sector
```

## d) Feature Selection
<a id="feature_selection"></a>

```python
""""
draw the distribution of repaid and cancelled loans together in one plot
add the mean of repaid and cancelled loans as vertical lines
"""

def feature_distribution(col, xlabel, repaid, cancelled):
    repaid_mean = np.mean(repaid)
    cancelled_mean = np.mean(cancelled)
    
    plt.figure(figsize=(6,7))
    sns.kdeplot(repaid, fill=True, label="repaid")
    sns.kdeplot(cancelled, fill=True, label="cancelled")
    plt.title("Distribution of: "+'"'+col+'"',fontsize=16)
    
    plt.axvline(x=repaid_mean, linestyle='--', linewidth=0.75)
    plt.axvline(x=cancelled_mean, linestyle='--', linewidth=0.75, color='coral')
    plt.xlabel(xlabel,fontsize=12)
    
    plt.show()


"""
conduct Mann-Whitney test on two distributions
Scale the data between (-1,+1): x_scaled = 2*(x - min(x_origin))/range(x_origin) - 1
"""

def mann_whitney(col, xlabel):
    repaid = df[df["loan_status"]==0][col]
    cancelled = df[df["loan_status"]==1][col]
    
    repaid_mean = np.mean(repaid)
    cancelled_mean = np.mean(cancelled)
    repaid_median = np.median(repaid)
    cancelled_median = np.median(cancelled)
    
    #print and plot
    print("Repaid - mean: ",repaid_mean,"/ median", repaid_median,"\n")
    print("Cancelled - mean: ",cancelled_mean,"/ median", cancelled_median,"\n")
    
    feature_distribution(col, xlabel, repaid, cancelled)
    
    # Mann-Whitney test
    p_value = sc.stats.mannwhitneyu(repaid, cancelled).pvalue
    
    print("P-value: ",p_value)
    if p_value<0.05:
        print("the distribution of: ", col," by loan repaid or cancelled are different at alpha=0.05")
    else:
        print("the distribution of: ", col," by loan repaid or cancelled are not different at alpha=0.05")
```

## e) Modeling and Evaluation
<a id="modeling_evaluation"></a>

```python
df.drop(columns=['borrower', 'project_name', 'first_repayment_date',  'last_repayment_date', 'agreement_signing_date', 'board_approval_date', 'effective_date', 'closed_date'], inplace=True)
ibrd_target_name = ['loan_status']
ibrd_feature_names = list(set(df.columns) - set(ibrd_target_name))
print("Final df columns")
pprint(list(df.columns))
```

```python
def balance_dataset(df,size):
    repaid_df = df[df['loan_status']==0]
    cancelled_df = df[df['loan_status']==1]
    
    repaid_idx = np.random.choice(list(repaid_df.index), int(len(cancelled_df)*size), replace=False)
    cancelled_idx= np.random.choice(list(cancelled_df.index), int(len(cancelled_df)*size), replace=False)
    
    repaid_balanced_df = df.loc[repaid_idx]
    cancelled_balanced_df = df.loc[cancelled_idx]
    
    balanced_df = pd.concat([repaid_balanced_df, cancelled_balanced_df])
    balanced_df = shuffle(balanced_df)
    
    return balanced_df

def split_x_y(df, features):
    categ_cols = df.select_dtypes(include = 'category').columns
    
    for col in categ_cols:
        le= ppr.LabelEncoder()
        le.fit(df[col])
        df[col]= le.transform(df[col])
        
        
    x= np.array(df[features])
    y= np.array(df['loan_status'])
    
    return x, y
```

```python
# balanced training set
train_df  = balance_dataset(df, 0.8)
print("Size of balanced training dataset:", len(train_df))
print("Count of cancelled loans in balanced training dataset:", 
     len(train_df[train_df['loan_status']==1]))

# training set x, y
train_x, train_y = split_x_y(train_df, ibrd_feature_names)
```

```python
# balanced validation set
valid_df = df.iloc[list(set(df.index)-set(train_df.index))]
# valid_df  = balance_dataset(valid_df, 1)

# print("Size of balanced validation dataset:", len(valid_df))
# print("Count of cancelled loans in balanced validation dataset:", len(valid_df[valid_df['loan_status']==1]))

# # validation set x, y
# valid_x, valid_y = split_x_y(valid_df, ibrd_feature_names)


# Ensure valid_df is not empty before balancing
if not valid_df.empty:
    valid_df = balance_dataset(valid_df, 1)
else:
    print("Warning: Validation dataset is empty. Cannot balance.")

# Print validation set size and counts
print("Size of balanced validation dataset:", len(valid_df))
print("Count of cancelled loans in balanced validation dataset:", len(valid_df[valid_df['loan_status'] == 1]))

# Validation set x, y
if not valid_df.empty:
    valid_x, valid_y = split_x_y(valid_df, ibrd_feature_names)
else:
    valid_x, valid_y = None, None  # Or handle it as needed
```

### &bull; Base Model

```python
def model_evaluation(clf, train_x, train_y, valid_x, valid_y):
    scores = cross_val_score(clf,train_x, train_y, cv=5, error_score='raise')
    clf.fit(train_x, train_y)
    pred = clf.predict(valid_x)
    
    print("Cross Validation Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
    # Confusion matrix, precision, recall, f-score
    print("\n Prediction against Validation Set\n - Confusion Matrix\n", confusion_matrix(valid_y, pred))
    print("\n- Classification Report\n", classification_report(valid_y, pred))
    print("\t (0: Repaid, 1: Cancelled)")
```
